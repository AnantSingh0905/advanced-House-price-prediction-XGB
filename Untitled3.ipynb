```python
import pandas as pd
import seaborn as sns
import numpy as np
```

```python
train = pd.read_csv('train.csv')
```

```python
test = pd.read_csv('test.csv')
```

```python
print(train.isnull().sum().sum())
print(test.isnull().sum().sum())
```

```python
final=pd.concat([train,test],axis=0)
```

```python
final.drop(columns=['PoolQC','Alley','MiscFeature','LotFrontage','MasVnrType','Fence','Id','FireplaceQu','SalePrice'],inplace=True)
```

```python
num_cols = final.select_dtypes(include=[np.number]).columns.tolist()
cat_cols = final.select_dtypes(exclude=[np.number]).columns.tolist()
```

```python
def fill_missing_values(df, cat_cols, num_cols):
    # Fill missing values for categorical columns with the mode
    for col in cat_cols:
        mode_value = df[col].mode()[0]
        df[col].fillna(mode_value, inplace=True)
    
    # Fill missing values for numerical columns with the mean
    for col in num_cols:
        med_value = df[col].median()
        df[col].fillna(med_value, inplace=True)
        
fill_missing_values(final, cat_cols, num_cols)
```

```python
final.isnull().sum().sum()
```

```python
dummy=pd.get_dummies(final[cat_cols],drop_first=True)
dummy.shape
```

```python
final.drop(columns=cat_cols,inplace=True)
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
scaled_numeric_features = scaler.fit_transform(final)
scaled_numeric_df = pd.DataFrame(scaled_numeric_features, columns=final.columns)
```

```python
# Step 1: Reset indices to ensure alignment
scaled_numeric_df = scaled_numeric_df.reset_index(drop=True)
dummy = dummy.reset_index(drop=True)

# Step 2: Concatenate the DataFrames
final = pd.concat([scaled_numeric_df, dummy], axis=1)
```

```python
final.tail
```

```python
scaled_numeric_df
```

```python
x_train=final.iloc[:1460,:]
x_test=final.iloc[1460:,:]
```

```python
y_train=train['SalePrice']
```

```python
param_grid = {
    'n_estimators': [500, 1000, 1500],
    'learning_rate': [0.01, 0.05, 0.1],
    'max_depth': [4, 6, 8], 
    'min_child_weight': [1, 3],
    'gamma': [0, 0.1, 0.3],
    'subsample': [0.7, 0.8, 0.9], 
    'colsample_bytree': [0.7, 0.8, 0.9], 
    'booster': ['gbtree'],
    'reg_alpha': [0, 0.1, 1],
    'reg_lambda': [0.1, 1, 10],
    'base_score': [0.5, 0.75],
}
```

```python
import xgboost as xgb
xgb_reg=xgb.XGBRegressor()
xgb_reg.fit(x_train,y_train) 
```

```python
from sklearn.model_selection import RandomizedSearchCV

rs = RandomizedSearchCV(
    estimator=xgb_reg,
    param_distributions=param_grid,
    n_iter=150,
    cv=5, 
    scoring='neg_root_mean_squared_error', 
    n_jobs=4, 
    verbose=5, 
    return_train_score=True, 
    random_state=42 
)
```

```python
rs.fit(x_train,y_train)

# Get the best parameters and corresponding score
best_params = rs.best_params_
best_score = rs.best_score_

print("Best Parameters:", best_params)
print("Best Score (Negative MSE):", best_score)
```
