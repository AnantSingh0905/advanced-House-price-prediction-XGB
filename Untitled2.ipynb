```python
from scipy import stats
import numpy as np

# Calculate the Z-scores for each feature
z_scores = np.abs(stats.zscore(numeric_train_data))

# Set a threshold for the Z-score, usually 3
threshold = 3

# Create a boolean mask for rows where any Z-score exceeds the threshold
outliers_mask = (z_scores > threshold).any(axis=1)

# Keep only the rows that are not outliers
Train_dataset = Train_dataset[~outliers_mask]

# Display the resulting dataframe without outliers
print("Data without outliers:", Train_dataset)
```

```python
Train_dataset
```

```python
# Separate features (X) and target (y)
X = Train_dataset.drop(columns=['SalePrice','LogSalePrice'])  # Drop 'SalePrice' and 'Id'
y = Train_dataset['LogSalePrice']  # Target variable
```

```python
from sklearn.decomposition import PCA
import numpy as np

# Apply PCA
pca = PCA(n_components=0.90)  # Retain 90% of the variance
X_pca = pca.fit_transform(X)

# Explained variance ratio
explained_variance = np.cumsum(pca.explained_variance_ratio_)
print("Explained Variance:", explained_variance)
```

```python
import xgboost as xgb
from sklearn.metrics import mean_squared_error
import numpy as np

# Convert training and validation data into DMatrix objects
dtrain = xgb.DMatrix(X_train, label=y_train)
dval = xgb.DMatrix(X_test, label=y_test)

# Set the parameters
params = {
    'eta': 0.04, 
    'max_depth': 6, 
    'subsample': 0.5,
    'colsample_bytree': 1, 
    'gamma': 0.1,
    'min_child_weight': 4,
    'objective': 'reg:squarederror',  # For regression tasks
    'eval_metric': 'rmse',
    'seed': 42
}

# Define the evaluation set
evals = [(dtrain, 'train'), (dval, 'eval')]

# Train the model using xgb.train and early stopping
xgb_model = xgb.train(
    params, 
    dtrain, 
    num_boost_round=500, 
    evals=evals, 
    early_stopping_rounds=50,  # Stop if no improvement for 50 rounds
    verbose_eval=False         # Suppress output
)

# Predict on validation set
y_pred_xgb = xgb_model.predict(dval)

# Calculate RMSE
rmse_xgb = np.sqrt(mean_squared_error(y_test, y_pred_xgb))
print("XGBoost RMSE with early stopping:", rmse_xgb)
```

```python
import xgboost as xgb
from sklearn.metrics import mean_squared_error
import numpy as np

# Convert training and validation data into DMatrix objects
dtrain = xgb.DMatrix(X_train, label=y_train)
dval = xgb.DMatrix(X_test, label=y_test)

# Set the parameters for slower learning
params = {
    'eta': 0.01,  # Lower learning rate (make it slow)
    'max_depth': 6, 
    'subsample': 0.5,
    'colsample_bytree': 1, 
    'gamma': 0.2,  # Slightly higher gamma to make it more conservative
    'min_child_weight': 6,  # Higher value to ensure leaves contain more weight
    'objective': 'reg:squarederror',  # For regression tasks
    'eval_metric': 'rmse',
    'seed': 42
}

# Define the evaluation set
evals = [(dtrain, 'train'), (dval, 'eval')]

# Train the model using xgb.train and early stopping
xgb_model = xgb.train(
    params, 
    dtrain, 
    num_boost_round=1000,  # Increase number of boosting rounds
    evals=evals, 
    early_stopping_rounds=50,  # Stop if no improvement for 50 rounds
    verbose_eval=False         # Suppress output
)

# Predict on validation set
y_pred_xgb = xgb_model.predict(dval)

# Calculate RMSE
rmse_xgb = np.sqrt(mean_squared_error(y_test, y_pred_xgb))
print("XGBoost RMSE with slower learning:", rmse_xgb)
```

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_squared_error
from sklearn.model_selection import cross_val_score
import numpy as np

# Initialize and train the Random Forest model with tuned parameters
rf_model = RandomForestRegressor(
    n_estimators=500,         # Increase number of trees for better averaging
    max_depth=30,             # Limit the depth of each tree to avoid overfitting
    min_samples_split=5,      # Min samples required to split an internal node
    min_samples_leaf=2,       # Min samples required to be at a leaf node
    max_features='sqrt',      # Use sqrt of features for better splits
    random_state=42
)

# Fit the model on training data
rf_model.fit(X_train, y_train)

# Predict on validation set
y_pred_rf = rf_model.predict(X_test)

# Calculate RMSE on validation data
rmse_rf = np.sqrt(mean_squared_error(y_test, y_pred_rf))
print("Updated Random Forest RMSE:", rmse_rf)

# Optional: Cross-Validation for more robust evaluation
cv_scores = cross_val_score(rf_model, X_train, y_train, cv=5, scoring='neg_mean_squared_error')
rmse_cv = np.sqrt(-cv_scores)
print("Cross-Validation RMSE scores:", rmse_cv)
print("Average CV RMSE:", np.mean(rmse_cv))
```

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_squared_error
from sklearn.model_selection import GridSearchCV, cross_val_score

# Initialize the base RandomForestRegressor
rf_model = RandomForestRegressor(n_estimators=100, random_state=42)

# Fit the model
rf_model.fit(X_train, y_train)

# Predict on test set
y_pred_rf = rf_model.predict(X_test)

# Calculate RMSE for the initial model
rmse_rf = np.sqrt(mean_squared_error(y_test, y_pred_rf))
print("Initial Random Forest RMSE:", rmse_rf)

# Displaying feature importance to understand which features are most important
feature_importances = rf_model.feature_importances_
important_features = pd.Series(feature_importances, index=X_train.columns).sort_values(ascending=False)
print("\nFeature Importances:\n", important_features)

# Hyperparameter tuning using GridSearchCV
param_grid = {
    'n_estimators': [100, 200, 300, 500],  # You can adjust this based on your computation capacity
    'max_depth': [10, 20, None],
    'min_samples_split': [2, 5, 10],
    'min_samples_leaf': [1, 2, 4],
    'max_features': ['auto', 'sqrt', 'log2'],
    'bootstrap': [True, False]
}

# Initialize GridSearchCV to tune the hyperparameters
grid_search = GridSearchCV(estimator=rf_model, param_grid=param_grid, cv=5, n_jobs=-1, verbose=2)
grid_search.fit(X_train, y_train)

# Get the best model from GridSearchCV
best_rf_model = grid_search.best_estimator_

# Predict using the best model
y_pred_best_rf = best_rf_model.predict(X_test)

# Calculate RMSE for the tuned model
rmse_best_rf = np.sqrt(mean_squared_error(y_test, y_pred_best_rf))
print("\nTuned Random Forest RMSE:", rmse_best_rf)

# Cross-validation to validate performance
cv_rmse = cross_val_score(best_rf_model, X_train, y_train, cv=5, scoring='neg_root_mean_squared_error')
print("\nCross-validated RMSE:", -cv_rmse.mean())

# If the RMSE is still not good enough, try increasing n_estimators further
```

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import GridSearchCV
from sklearn.metrics import mean_squared_error
import numpy as np

# Define an expanded parameter grid for Random Forest
param_grid = {
    'n_estimators': [100, 200, 300, 500],        # Increasing number of trees
    'max_depth': [10, 20, 30, None],             # Add None for no depth limit
    'min_samples_split': [2, 5, 10],             # Minimum samples required to split
    'min_samples_leaf': [1, 2, 4],               # Minimum samples required at a leaf node
    'max_features': ['auto', 'sqrt', 'log2']     # Number of features to consider for the best split
}

# Perform grid search with cross-validation (5-fold)
grid_search = GridSearchCV(
    estimator=RandomForestRegressor(random_state=42), 
    param_grid=param_grid, 
    cv=5, 
    scoring='neg_mean_squared_error', 
    verbose=1,        # Display progress during fitting
    n_jobs=-1         # Use all processors
)

# Fit the grid search to the data
grid_search.fit(X_train, y_train)

# Extract the best model and its parameters
best_rf_model = grid_search.best_estimator_
print("Best Random Forest parameters:", grid_search.best_params_)

# Evaluate the best model on the validation set
y_pred_best_rf = best_rf_model.predict(X_test)
rmse_best_rf = np.sqrt(mean_squared_error(y_test, y_pred_best_rf))
print("Best Random Forest RMSE:", rmse_best_rf)
```
